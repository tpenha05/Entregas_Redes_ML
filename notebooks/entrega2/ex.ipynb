```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
mean_0 = np.array([1.5, 1.5])
cov_0 = np.array([[0.5, 0], [0, 0.5]])

mean_1 = np.array([5, 5])
cov_1 = np.array([[0.5, 0], [0, 0.5]])

n_samples = 1000

x_0 = np.random.multivariate_normal(mean_0, cov_0, n_samples)
x_1 = np.random.multivariate_normal(mean_1, cov_1, n_samples)
```

```python
plt.scatter(x_0[:, 0], x_0[:, 1], color='blue', alpha=0.5, label='Class 0')
plt.scatter(x_1[:, 0], x_1[:, 1], color='red', alpha=0.5, label='Class 1')
plt.title('Synthetic Data Distribution')
plt.xlabel('Feature 1') 
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.vstack((x_0, x_1))
y = np.array([0]*n_samples + [1]*n_samples)

w = np.array([0,0])
b = 0
lr = 0.01

accuracies = {}
dict_errors = {}
```

```python
for epoch in range(100):
    for n in range(len(x)):
        z = (x[n] @ w) + b
        y_pred = 1 if z >= 0 else 0
        error = y[n] - y_pred
        if error != 0:
            w = w + lr * error * x[n]
            b = b + lr * error
    y_pred_all = np.where((x @ w + b) >= 0, 1, 0)
    errors_epoch = np.sum(y_pred_all != y)
    accuracies[epoch] = 1 - errors_epoch/len(x)
    if errors_epoch == 0:
        print(f"Convergiu em {epoch} pocas")
        break

if errors_epoch != 0:
    print(f"No convergiu. Acurcia final: {accuracies[epoch]}")
```

```python
plt.scatter(x_0[:, 0], x_0[:, 1], color='blue', alpha=0.5, label='Class 0')
plt.scatter(x_1[:, 0], x_1[:, 1], color='red', alpha=0.5, label='Class 1')
x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
x_vals = np.linspace(x_min, x_max, 100)
y_vals = -(w[0] * x_vals + b) / w[1]

plt.plot(x_vals, y_vals, color='black', label='Decision boundary')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(list(accuracies.keys()), list(accuracies.values()), marker='o', color='blue')
plt.xlabel("pocas")
plt.ylabel("Acurcia")
plt.title("Evoluo da Acurcia por poca")
plt.grid(True)
plt.show()
```
